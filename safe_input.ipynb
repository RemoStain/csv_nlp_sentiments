### `from getpass import getpass`

**Purpose:**  
Import a function that reads user input without displaying it on the screen.

**Behavior:**
- Replaces normal `input()` when sensitive data is entered.
- Characters typed by the user are hidden.
- Commonly used for passwords or secret tokens.

**Effect:**  
Prevents sensitive information from being visible in the terminal.

**Plain meaning:**  
Ask the user for text without showing what they type.

```python
def safe_input(expected_type, message=None, default=None, is_password:bool=False):
    """
    Prompt the user for input and safely convert it to the given type.

    Args:
        expected_type (type): The type to convert the input into (e.g., int, float, str, bool).
        message (str, optional): Custom message to display instead of the default.
        default (any, optional): Default value to return if conversion fails.
        is_password (bool, optional): If True, input will be hidden (for passwords).

    Returns:
        The converted value of the correct type, or None if interrupted.
    """
    while True:
        try:
            prompt = (
                message
                if message is not None
                else f"Enter a(n) ({expected_type.__name__}): "
            )
            if is_password:
                user_input = getpass(prompt)
            else:
                user_input = input(prompt)

            # Special handling for bool
            if expected_type is bool:
                lowered = user_input.strip().lower()
                if lowered in ("true", "1", "yes", "y"):
                    return True
                elif lowered in ("false", "0", "no", "n"):
                    return False
                else:
                    raise ValueError("Invalid boolean input")

            if expected_type is str:
                if user_input == "" and default is not None:
                    return default

            # Attempt conversion for other types
            return expected_type(user_input)

        except ValueError:
            if default is not None:
                if user_input != "":
                    print(f"Invalid input. Using default value: {default}")
                return default
            print(f"Invalid input. Please enter a valid {expected_type.__name__}.")

        except KeyboardInterrupt:
            print("\nKeyboard interrupt detected. Aborting input.")
            return None  # or raise SystemExit if you want to exit the program
        except EOFError:
            print("\nEnd of input detected (Ctrl+D). Returning None.")
            return None

```

`safe_input(expected_type, message=None, default=None, is_password: bool = False)`

**Purpose:**  
Safely ask the user for input and convert it to a specific type without crashing the program.

**Behavior:**
- Displays a prompt asking the user for input.
- Uses a custom message if provided; otherwise builds one from `expected_type`.
- Hides input if `is_password` is `True`.
- Repeats until valid input is received or the user interrupts.
- Converts the input into `expected_type`.
- Handles `bool` specially, accepting values like `yes/no`, `true/false`, `1/0`.
- If conversion fails and a `default` is provided, returns that default.
- Gracefully handles `Ctrl+C` and `Ctrl+D` by returning `None`.

**Effect:**  
Prevents invalid user input from crashing the program and guarantees a predictable return value.

**Plain meaning:**  
Ask the user for something, keep asking until it makes sense, and never let bad input break the program.

---

### Parameter breakdown

`expected_type`

**Purpose:**  
Define what type the input should become.

**Behavior:**  
The function attempts to convert the user’s input into this type.

**Effect:**  
Controls how the raw text is interpreted.

**Plain meaning:**  
“What kind of value do you want? An `int`, `float`, `str`, or `bool`?”

---

`message`

**Purpose:**  
Customize what the user sees as the prompt.

**Behavior:**  
If provided, it replaces the default prompt text.

**Effect:**  
Allows clearer or more specific instructions.

**Plain meaning:**  
The sentence shown to the user.

---

`default`

**Purpose:**  
Provide a fallback value when input is invalid.

**Behavior:**  
If conversion fails and `default` is set, the function returns it instead of retrying forever.

**Effect:**  
Keeps the program moving forward even with bad input.

**Plain meaning:**  
“What value should be used if the user messes up?”

---

`is_password`

**Purpose:**  
Control whether the input is visible while typing.

**Behavior:**  
If `True`, uses `getpass()` so characters are hidden.

**Effect:**  
Protects sensitive input like passwords.

**Plain meaning:**  
“Should the user’s typing be hidden from the screen?”

---

### Overall workflow

1. Build a prompt message.
2. Ask the user for input (hidden if `is_password` is true).
3. If the expected type is `bool`, interpret common true/false words.
4. If the expected type is `str` and the input is empty, return `default` if provided.
5. Attempt to convert the input to `expected_type`.
6. If conversion fails:
   - Return `default` if one exists.
   - Otherwise, print an error and ask again.
7. If the user presses `Ctrl+C` or `Ctrl+D`, stop and return `None`.

Result:  
The function always returns a valid value of the requested type, a default value, or `None` if input is aborted.


```python
# Example usage
age = safe_input(int, "Enter your age: ", default=18)
print(f"Your age is: {age}")
height = safe_input(float, "Enter your height in meters: ", default=1.75)
print(f"Your height is: {height} meters")
name = safe_input(str, "Enter your name: ", default="Guest")
print(f"Hello, {name}!")
wants_newsletter = safe_input(bool, "Do you want to subscribe to the newsletter? (yes/no): ", default=False)
print(f"Newsletter subscription: {wants_newsletter}")
password = safe_input(str, "Enter your password: ", is_password=True)
print(f"Your password is: {'*' * len(password) if password else 'None'}")
```
